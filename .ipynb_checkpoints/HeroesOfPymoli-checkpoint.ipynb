## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#unique_id = purchase_data[["Gender","SN"]].drop_duplicates()
bins = [0,9,14,19,24,29,34,99]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34",">35"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels = age_labels)
purchase_data[["Age Group","SN"]].drop_duplicates()
total_group = purchase_data.groupby(["Age Group"]).count()
age_range = purchase_data[["Age Group","SN"]].drop_duplicates()
#age_range
count_group = age_range["Age Group"].value_counts()
group_perc = ((count_group / unique_players) * 100).map(("{:.2f}%").format)
age_df = pd.DataFrame({"Age Group Count": count_group,
                      "Percent": group_perc})
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

bins = [0,9,14,19,24,29,34,99]
age_labels = ["<10","10-14","15-19","20-24","25-29","30-34",">35"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels = age_labels)

count_by_age = purchase_data.groupby("Age Group").count()["Purchase ID"]
avg_cost_by_age = purchase_data.groupby("Age Group").mean()["Price"].map("${:.2f}".format)
total_price_by_age = purchase_data.groupby("Age Group").sum()["Price"].map("${:.2f}".format)
# avg_cost_per_person = purchase_data.groupby("Age Group").mean()["Price"]

#avg_price_by_age
#age_df = pd.DataFrame({"Age Group":purchase_data["Age Group"], 
 #                            "Average Price": avg_price_by_age, 
  #                           "Purchase Count":purchase_count_by_age,
   #                         "Total Purchase":total_price_by_age})
#count_by_age
#avg_cost_by_age
#total_price_by_age
#age_group
age_df = pd.DataFrame({
                      "Purchase Count": count_by_age,
                      "Average Price": avg_cost_by_age,
                      "Total Price":total_price_by_age
})
                     
age_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pop_items_count = purchase_data.groupby(["Item ID", "Item Name"])["Price"].count()
pop_items_sales = purchase_data.groupby(["Item ID", "Item Name"])["Price"].sum()
item_total_sales = pop_items_sales / pop_items_count

pop_items_df = pd.DataFrame({"Count Purchased": pop_items_count,
                            "Total Sales":pop_items_sales,
                            "Price per Item": item_total_sales})
pop_items_df.sort_values(by = ["Count Purchased"], inplace = True, ascending = False)
pop_items_df.head()
```
